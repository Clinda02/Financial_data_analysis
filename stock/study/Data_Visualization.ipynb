# **기본함수**

```python
import matplotlib.pyplot as plt

dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]

plt.figure(figsize=(10, 6))
plt.title("Stock Prices Over 5 Days")
plt.xlabel("Date")
plt.ylabel("Price")
plt.plot(dates, prices, marker='o', label='Stock Price', color='blue')
plt.legend()
plt.grid()
plt.ylim(95, 110)
plt.xlim(0, 4)
```

```python
# 다중 플롯
plt.subplot(1, 2, 1)
plt.plot(dates, prices, marker='o', color='green')
plt.title("Stock Prices - Subplot 1")
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
plt.bar(dates, prices, color='orange')
plt.title("Stock Prices - Subplot 2")
plt.xticks(rotation=45)

plt.show()
plt.savefig("stock_prices_5_days.png")
```

# **막대 그래프(matplotlib)**

```python
import matplotlib.pyplot as plt
import numpy as np

dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
stock_A = [100, 102, 101, 105, 103]
stock_B = [95, 105, 102, 85, 101]
stock_C = [99, 101, 100, 107, 104]

x = np.arange(len(dates))
width = 0.25

plt.bar(x - width, stock_A, width, label='Stock A', color='blue') # hatch="/"
plt.bar(x, stock_B, width, label='Stock B', color='orange')
plt.bar(x + width, stock_C, width, label='Stock C', color='green')

plt.title("Stock Prices Comparison Over 5 Days")
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(x, dates)
plt.legend()
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

# **막대 그래프(SNS)**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
stock_A = [100, 102, 101, 105, 103]
stock_B = [95, 105, 102, 85, 101]
stock_C = [99, 101, 100, 107, 104]

data = pd.DataFrame({
    'Date': dates,
    'Stock A': stock_A,
    'Stock B': stock_B,
    'Stock C': stock_C
})

# Melt the DataFrame to long format for seaborn
data_melted = data.melt(id_vars='Date', var_name='Stock', value_name='Price')

plt.figure(figsize=(10, 6))
sns.barplot(x='Date', y='Price', hue='Stock', data=data_melted, palette=['blue', 'orange', 'green'])

plt.title("Stock Prices Comparison Over 5 Days")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

# **막대 그래프(Plotly)**

```python
import plotly.graph_objects as go

dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
stock_A = [100, 102, 101, 105, 103]
stock_B = [95, 105, 102, 85, 101]
stock_C = [99, 101, 100, 107, 104]

# 막대의 폭을 계산합니다
width = 0.25
x = [0, 1, 2, 3, 4]  # x축의 인덱스

fig = go.Figure()

fig.add_trace(go.Bar(x=[i - width for i in x], y=stock_A, name='Stock A', marker_color='blue'))
fig.add_trace(go.Bar(x=x, y=stock_B, name='Stock B', marker_color='orange'))
fig.add_trace(go.Bar(x=[i + width for i in x], y=stock_C, name='Stock C', marker_color='green'))

fig.update_layout(
    title="Stock Prices Comparison Over 5 Days",
    xaxis_title="Date",
    yaxis_title="Price",
    barmode='group',
    xaxis=dict(tickvals=x, ticktext=dates),
    template="plotly_white"
)

fig.show()
```

# **산점도 그래프(SNS)**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

prices = [100, 102, 101, 105, 103, 107, 110, 108, 111, 115]
volumes = [200, 220, 180, 250, 230, 270, 300, 290, 310, 320]
returns = [20, 10, -20, 30, -10, 20, 40, -10, 30, 50]

data = pd.DataFrame({
    'Stock Price': prices,
    'Volume': volumes,
    'Returns': returns
})

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Stock Price', y='Volume', color='blue', s=100, label='Price vs Volume')
sns.scatterplot(data=data, x='Stock Price', y='Returns', color='orange', s=100, label='Price vs Returns (Scaled)')
plt.title("Scatter Plot of Stock Prices, Volume, and Returns")
plt.xlabel("Stock Price")
plt.ylabel("Volume / Returns")
plt.legend()
plt.grid()
plt.show()
```

# **산점도 그래프(Plotly)**

```python
import plotly.graph_objects as go

prices = [100, 102, 101, 105, 103, 107, 110, 108, 111, 115]
volumes = [200, 220, 180, 250, 230, 270, 300, 290, 310, 320]
returns = [20, 10, -20, 30, -10, 20, 40, -10, 30, 50]

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=prices,
    y=volumes,
    mode='markers',
    marker=dict(color='blue', size=10),
    name='Price vs Volume'
))

fig.add_trace(go.Scatter(
    x=prices,
    y=returns,
    mode='markers',
    marker=dict(color='orange', size=10),
    name='Price vs Returns (Scaled)'
))

fig.update_layout(
    title="Scatter Plot of Stock Prices, Volume, and Returns",
    xaxis_title="Stock Price",
    yaxis_title="Volume / Returns",
    legend=dict(x=0.1, y=0.9),
    template="plotly_white"
)

fig.show()
```

# **히스토그램(Plotly)**

```python
import plotly.express as px
import numpy as np

stock_prices = np.random.normal(loc=100, scale=10, size=1000)

fig = px.histogram(stock_prices, nbins=30,
                   labels={'value': 'Price'},
                   title='Stock Price Distribution',
                   color_discrete_sequence=['blue'])

fig.update_traces(marker=dict(line=dict(color='black', width=1)))
fig.update_layout(xaxis_title='Price', yaxis_title='Frequency',
                  template='plotly_white')

fig.show()
```

# **파이차트(matplotlib)**

```python
import matplotlib.pyplot as plt

labels = ['Stock A', 'Stock B', 'Stock C', 'Stock D']
sizes = [40, 30, 20, 10]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.title("Stock Portfolio Distribution")
plt.show()
```

# **파이차트(Plotly)**

```python
import plotly.graph_objects as go

labels = ['Stock A', 'Stock B', 'Stock C', 'Stock D']
sizes = [40, 30, 20, 10]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = [0.1, 0, 0, 0]

fig = go.Figure(data=[go.Pie(labels=labels, values=sizes,
                               textinfo='label+percent',
                               pull=explode,
                               marker=dict(colors=colors))])

fig.update_layout(title="Stock Portfolio Distribution")
fig.show()
```

# **다중플롯(SNS)**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
stock_a = [100, 102, 101, 105, 103]
stock_b = [98, 99, 97, 100, 98]

data = pd.DataFrame({
    'Date': dates * 2,
    'Price': stock_a + stock_b,
    'Stock': ['Stock A'] * len(stock_a) + ['Stock B'] * len(stock_b)
})

g = sns.FacetGrid(data, col='Stock', height=5, aspect=1)
g.map(sns.lineplot, 'Date', 'Price', marker='o')

g.set_titles(col_template="{col_name} Prices")
g.set_axis_labels("Date", "Price")

g.axes[0, 0].set_ylim(95, 110)
g.axes[0, 1].set_ylim(95, 105)

g.tight_layout()
plt.show()
```

# **다중플롯(Plotly)**

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
stock_a = [100, 102, 101, 105, 103]
stock_b = [98, 99, 97, 100, 98]


fig = make_subplots(rows=1, cols=2, subplot_titles=("Stock A Prices", "Stock B Prices"))

fig.add_trace(go.Scatter(x=dates, y=stock_a, mode='lines+markers', name='Stock A Prices',
                         marker=dict(color='blue')), row=1, col=1)

fig.add_trace(go.Scatter(x=dates, y=stock_b, mode='lines+markers', name='Stock B Prices',
                         marker=dict(color='red')), row=1, col=2)

fig.update_layout(title="Stock Prices",
                  xaxis_title="Date",
                  yaxis_title="Price",
                  height=600)

fig.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 삼성전자와 카카오 주가 데이터
samsung = yf.Ticker('005930.KS')  # 삼성전자 종목 코드
kakao = yf.Ticker('035720.KS')    # 카카오 종목 코드

# 최근 1년간 두 종목 데이터
samsung_data = samsung.history(period='1y')
kakao_data = kakao.history(period='1y')
```

**matplotlib 사용하여 라인그래프 비교**

```python
plt.figure(figsize=(12, 6))
plt.plot(samsung_data['Close'], label='Samsung Electronics', color='blue')
plt.plot(kakao_data['Close'], label='Kakao', color='orange')

plt.title('Comparison of Stock Prices: Samsung Electronics vs Kakao (Last 3 Years)')
plt.xlabel('Date')
plt.ylabel('Closing Price (KRW)')
plt.legend()
plt.grid()
plt.show()
```
